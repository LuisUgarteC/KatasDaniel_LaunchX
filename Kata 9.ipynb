# Ejercicio 1 - Argumentos en Funciones

```python
def reporte(tanque_principal, tanque_secundario, tanque_extra):
    Promedio_Total = (tanque_principal + tanque_secundario + tanque_extra) / 3
    return f"""Reporte:
    Promedio: {Promedio_Total}%
    Tanque principal: {tanque_principal}%
    Tanque secundario: {tanque_secundario}%
    Tanque extra: {tanque_extra}% 
    """

print(reporte(40, 100, 95))

def promedio(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

promedio([80, 85, 81]) 

def reporte(tanque_principal, tanque_secundario, tanque_extra):
    return f"""Reporte:
    Promedio: {promedio([tanque_principal, tanque_secundario, tanque_extra])}%
    Tanque principal: {tanque_principal}%
    Tanque secundario: {tanque_secundario}%
    Tanque extra: { tanque_extra}% 
    """

print(reporte(70, 66, 69))
```

# Ejercicio 2 - Argumentos Palabras Clave

```python
def mision(tiempo_lanzamiento, tiempo_vuelo, destino, tanque_secundario, tanque_principal):
    return f"""
    Mision {destino}
    Tiempo total: {tiempo_lanzamiento + tiempo_vuelo} minutos
    Combustible total: {tanque_secundario + tanque_principal} litros
    """

print(mision(8, 25, "Luna", 660000, 450000))

def mision(destino, *minutos, **combustible):
    return f"""
    Mision {destino}
    Tiempo: {sum(minutos)} minutos
    Combustible: {sum(combustible.values())}
    """

print(mision("Luna", 15, 25, 81, principal=4200000, secundario=6900000))

def mision(destino, *minutos, **combustible):
    reporte = f"""
    Mision {destino}
    Tiempo de viaje: {sum(minutos)} minutos
    Combustible total: {sum(combustible.values())}
    """
    for tanque, litros in combustible.items():
        reporte += f"{tanque} {litros} litros\n"
    return reporte

print(mision("Luna", 20, 21, 95, principal=660000, secundario=600000))
```
