# Ejercicio 1 - Listas de Python

Información Planetaria


```python
planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Neptuno']
print(planets)

number_planets = len(planets)
print('Existen', number_planets, '\n')

planets.append('Plutón')
number_planets = len(planets)

print('Se agregó Plutón a la lista. El último planeta es: ', planets[-1])
print(planets)
print('Existen', number_planets)
```

# Ejercicio 2 - Datos de una Lista

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']

planeta = input("Escoge un Planeta: ")

planet_index = planets.index(planeta)

print('Planetas más cercanos a ' + planeta)
print(planets[0:planet_index])

print('Planetas más lejanos a ' + planeta)
print(planets[planet_index + 1:])
```
