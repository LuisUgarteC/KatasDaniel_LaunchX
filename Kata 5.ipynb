# Ejercicio 1 - Operadores Aritméticos

Distancia entre dos planetas

```python
Tierra = 149597870
Jupiter = 778547200

Distancia = Tierra - Jupiter

print("La distancia estre La tierra y Júpiter es: ")
print(abs(Distancia), "km") 

Distancia_Millas = Distancia * 0.621
print("\nLa distancia anterior en millas es: ")
print(round(abs(Distancia_Millas)))
```
